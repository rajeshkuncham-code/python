```python
from scipy.stats import zscore

data=sns.load_dataset('iris')
df=data.copy()
```

```python
z_score = np.abs(zscore(df.drop('species',axis=1)))
z_score
```

```python
non_outliers = (z_score <=3).all(axis=1)
df_no_outliers = df[non_outliers]
```
